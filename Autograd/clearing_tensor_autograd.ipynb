```python
import pandas as pd 
import numpy as np 
import torch 
import torch.nn as  nn
from sklearn.model_selection import train_test_split
```

```python
X=torch.tensor([[4311,3123,1233],[4144,1335,6134]],requires_grad=True,dtype=torch.float32)
y=torch.tensor([[0.1,0.2,0.3]])
```

```python
w=torch.tensor([1,1,1],requires_grad=True,dtype=torch.float32)
b=torch.tensor(0.0,requires_grad=True)
```

```python
z=w*X+b 
z
```

```python
p=torch.sin(z)
p 
```

```python
sigmoid=nn.Sigmoid()
values=sigmoid(p)
values 
```

```python
loss=nn.MSELoss()
loss_v=loss(values,y)
loss_v 
```

```python
loss_v.backward()
```
