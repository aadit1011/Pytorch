```python
import time
```

```python
d1=torch.ones(30,30,dtype=torch.float32)
d1
```

```python
p1=torch.full((30,30),5,dtype=torch.float32)
p1
```

```python
start=time.time()
torch.matmul(p1,d1)
end=time.time()
print(start,end)
t=end-start
print('Total time on CPU.',t)
```

```python
d1=d1.to(device)
p1=p1.to(device)
start=time.time()
torch.matmul(p1,d1)
end=time.time()
print(start,end)
t2=end-start
print('Total time on CPU.',t2)
```

```python
t/t2
```
