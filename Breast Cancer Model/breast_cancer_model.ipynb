```python
import pandas as pd 
import numpy as np 
import torch 
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.metrics import accuracy_score

```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
df 
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['diagnosis']),df['diagnosis'],test_size=0.2,random_state=23,shuffle=True)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
label=LabelEncoder()
y_train=label.fit_transform(y_train)
y_test=label.transform(y_test)
```

```python
X_train_ten=torch.tensor(X_train,requires_grad=True)
X_test_ten=torch.tensor(X_test,requires_grad=True)
y_train_ten=torch.tensor(y_train,dtype=torch.float64)
y_test_ten=torch.tensor(y_test,dtype=torch.float64)
```

```python
class MyNeuralNetwork():
    np.random.seed(42)
    def __init__(self,X): 
        self.weights=torch.rand(X.shape[1],1,requires_grad=True,dtype=torch.float64)
        self.bias=torch.zeros(1,requires_grad=True,dtype=torch.float64) 

    def forward(self,X): 
        z=torch.matmul(X,self.weights) + self.bias
        y_pred=torch.sigmoid(z)
        return y_pred 

    def loss(self,y_pred,y): 
        bce=nn.BCELoss()
        los=bce(y_pred,y)
        return los
    
    def backpropagation(self,loss): 
        loss.backward()

    def update_weights(self): 
        with torch.no_grad():
            self.weights-=lr*self.weights.grad
            self.bias-=lr*self.bias.grad
        self.weights.grad.zero_()
        self.bias.grad.zero_()
        

    
        
```

```python
model=MyNeuralNetwork(X_train_ten)
model.weights
```

```python
for epoch in range(epochs): 
    y_pred=model.forward(X_train_ten)
    loss=model.loss(y_pred,y_train_ten.reshape(-1,1))
    print(f'Loss for Epoch {epoch+1} Loss {loss}')
    model.backpropagation(loss) 
    model.update_weights()
    
```
